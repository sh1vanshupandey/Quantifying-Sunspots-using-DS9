# Quantifying Sunspots using DS9






# Objective

The objective of this experiment is to analyze solar FITS images using SAOImage DS9 software and calculate the physical area of three distinct sunspots. The measured pixel areas are converted into physical units using the known solar radius.



# Software and Data Used

- SAOImage DS9
- Solar FITS files provided in assignment
- FITS files used:
  - 7 april 1 (around 11-30).fit
  - 7 april 2.fit
  - 7 april 3.fit
  - 7 april 4.fit



# Methodology

The FITS solar images were opened in SAOImage DS9 software. Three prominent sunspots were selected from the solar disk. Preference was given to isolated and clearly visible sunspots to reduce uncertainty caused by overlapping regions.

The ellipse region tool in DS9 was used to approximate the shape of each sunspot. The semi-major axis and semi-minor axis of each ellipse were measured in pixels. The pixel area of the sunspots was then calculated using the ellipse area formula:

$$A_{\text{pix}} = πab$$
where:
- a = semi-major axis in pixels
- b = semi-minor axis in pixels

To convert pixel measurements into physical units, the solar disk radius was measured in pixels using the circle region tool. Using the known solar radius:

$$R_{\text{sun}} = 696000 km$$

# Solar Radius Measurement


Measured solar radius in image:

$$
R_{\text{pix}} = 950 \ \text{pixels}
$$

Known solar radius:

$$
R_{\odot} = 696000 \ \text{km}
$$

Image scale:

$$
s = \frac{R_{\odot}}{R_{\text{pix}}}
$$

Substituting the values:

$$
s = \frac{696000}{950}
$$

$$
s \approx 732.6 \ \text{km/pixel}
$$

---


the image scale in km/pixel was calculated.

Finally, the physical area of each sunspot was obtained by multiplying the pixel area by the square of the image scale.

Potential errors include:
- blurred boundaries of sunspots
- irregular shapes
- manual fitting uncertainty
- projection effects near solar limb

## Sunspot 1


### Measured Parameters

- Semi-major axis $(a)$ = 21 pixels
- Semi-minor axis $(b)$ = 17.5 pixels

### Pixel Area Calculation

The pixel area of the sunspot is calculated using the ellipse area formula:

$$
A_{\text{pix}} = \pi ab
$$

Substituting the measured values:

$$
A_{\text{pix}} = \pi \times (21) \times (17.5)
$$

$$
A_{\text{pix}} \approx 1154.54 \ \text{pixels}^2
$$

---

### Physical Area Calculation

The physical area is calculated using:

$$
A_{\text{phys}} = A_{\text{pix}} \times s^2
$$

where:

$$
s = 732.6 \ \text{km/pixel}
$$

Substituting the values:

$$
A_{\text{phys}} = 1154.54 \times (732.6)^2
$$

$$
A_{\text{phys}} \approx 6.20 \times 10^8 \ \text{km}^2
$$

# Discussion

The measured sunspots showed significant variation in size and shape. Larger sunspots were easier to measure accurately, while smaller and faint sunspots introduced higher uncertainty. Sunspots near the solar limb appeared distorted because of projection effects, which could affect the calculated area.

The ellipse approximation worked reasonably well for most sunspots, although many spots were irregular in shape. Manual region selection and image resolution limitations were the primary sources of uncertainty.

# Conclusion

Solar FITS images were successfully analyzed using DS9 software. The areas of three distinct sunspots were measured in pixel units and converted into physical units using the known solar radius. The experiment demonstrated how astronomical image analysis can be used to estimate physical properties of solar features.
